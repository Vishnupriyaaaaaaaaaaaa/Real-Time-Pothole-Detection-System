```python
from ultralytics import YOLO
import cv2
#pipeline for preprocessing
def preprocess_frame(frame_bgr):
    frame = cv2.GaussianBlur(frame_bgr, (5, 5), 0)

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    v_eq = clahe.apply(v)

    hsv_eq = cv2.merge((h, s, v_eq))
    frame_eq = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2BGR)

    frame_rgb = cv2.cvtColor(frame_eq, cv2.COLOR_BGR2RGB)
    return frame_rgb

#camera activation
model = YOLO("runs/detect/train13/weights/best.pt")
cap = cv2.VideoCapture(1)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    processed = preprocess_frame(frame)

    results = model(processed, imgsz=640, verbose=False)

    annotated = results[0].plot()
    cv2.imshow("Pothole Detection with Preprocessing", annotated)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2


model = YOLO("runs/detect/train13/weights/best.pt")

input_path = "testimages/test7pothole.jpg"

def preprocess_frame(frame_bgr):
    frame = cv2.GaussianBlur(frame_bgr, (3, 3), 0)


    
    hsv  = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    v_eq  = clahe.apply(v)

    hsv_eq    = cv2.merge((h, s, v_eq))
    frame_eq  = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2BGR)

    return frame_eq 

image = cv2.imread(input_path)

processed = preprocess_frame(image)

results = model(processed, imgsz=640, conf=0.5, save=True)

print("Detection complete. Annotated result saved in 'runs/detect/predict' folder.")
```
